```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
df = pd.read_csv('https://github.com/mwcoyne2/BDI-Final-Project/raw/main/2021%20College%20Basketball.csv')
display(df)
```

```python
#Shape of Data Set
df.shape
```

```python
#I renamed columns so they are more user friendly for non basktball fans
df.rename(columns =
    {'TEAM':'Team',
    'CONF':'Conference',
    'G':'Games',
    'W':'Wins',
    'ADJOE':'Adj Offensive Efficiency',
    'ADJDE':'Adj Defensive Efficiency',
    'BARTHAG':'Power Ranking',
    'EFG_O':'Effective Field Goal Percentage Shot',
    'EFG_D':'Effective Field Goal Percentage Allowed',
    'TOR':'Turnover Rate',
    'TORD':'Steal Rate',
    'ORB':'Offensive Rebounds',
    'DRB':'Defensive Rebounds',
    'FTR':'Freee Throw Rate',
    'FTRD':'Free Throw Rate Allowed',
    '2P_O':'Two Point Percentage',
    '2P_D':'Two Point Percentage Allowed',
    '3P_O':'Three Point Percentage',
    '3P_D':'Three Point Percentage Allowed',
    'ADJ_T':'Adjusted Tempo',
    'WAB':'Wins Above Bubble',
    'POSTSEASON':'Postseason',
    'SEED':'Seed'},
    inplace = True
    )
df
```

```python
#Information on the data set
df.info()
```

```python
#Sample of Data
display(df.head())
display(df.tail())
display(df.sample())
```

```python
#Average Seed of conferences
df_average_conference_seed = df.groupby('Conference', as_index=False).agg({'Seed':['count','mean','median','max','min']})
df_average_conference_seed.sort_values([('Seed','count'),('Seed','min')], ascending=[False,True], inplace=True)
df_average_conference_seed
```
