#Offensive Statistics Analysis
**What I did:** The first thing I wanted to analyze is the effect Offens has on wininng. To to this I graphed the offense efficiency to wins as well as graphed a bar chart of three point percentage. I took the average of each of these metrics to make it easier and compare conferences better.

```python
df.head()
```

```python
df_average_offensive_stats = df.groupby('Conference', as_index=False).agg({
    'Wins':'mean',
    'Adj Offensive Efficiency':'mean',
    'Effective Field Goal Percentage Shot':'mean',
    'Offensive Rebounds':'mean',
    'Two Point Percentage':'mean',
    'Three Point Percentage':'mean'})
df_average_offensive_stats = df_average_offensive_stats.sort_values('Wins', ascending = False)
df_average_offensive_stats
```

```python
fig = px.bar(
    df_average_offensive_stats,
    x = 'Conference',
    y ='Adj Offensive Efficiency',
    title = 'Offensive Efficiency by Conference',
    color = 'Conference'
)
fig.show()
```

```python
px.scatter(
    df_average_offensive_stats,
    x = 'Adj Offensive Efficiency',
    y = 'Wins',
    title = 'Is there a correlatuon between Two point and Three Point Percentage?',
    text = 'Conference',
    color = 'Conference'
)
```

```python
fig = px.bar(
    df_average_offensive_stats,
    x = 'Conference',
    y = 'Three Point Percentage',
    title = 'Conference With The Best Three Point Percentage',
    color = 'Conference',
    
)
fig.show()
```

#Average Seed Analysis
**What I did:** The third thing I wanted to analyze was the tournament results of the conference. I used sorting and group by to help me analyze the tournement seeding of each conference respectively. I also decided to filter out all of the conferences with only 1 team in the tournement because those conferences are gaurenteed 1 team who is the conference winner

```python
#Average Seed of conferences
df_average_conference_seed = df.groupby('Conference', as_index=False).agg({'Seed':['count','mean','median','max','min']})
df_average_conference_seed.columns = ['Conference','Total Seeds', 'Seed Mean', 'Seed Median', 'Seed Max', 'Seed Min']
df_average_conference_seed.sort_values(['Total Seeds', 'Seed Min'], ascending=[False,True], inplace=True)
df_average_conference_seed = df_average_conference_seed[df_average_conference_seed['Total Seeds'] > 1]
df_average_conference_seed
```

```python
fig = px.box(df_average_conference_seed, x = 'Seed Mean', title = 'Average Seed Box Plot of Conferences with more then one Team')
fig.show()
```

```python
fig = px.bar(df_average_conference_seed, x = 'Conference', y = 'Total Seeds', title = 'Total Seeds by Conference With More Then One Bid', color = 'Conference')
fig.show()
```

**Findings:**when analyzing how conferences were seeded in the tournement I excluded conferences with only 1 team to have a better understanding of of how the teams are actually seeded. If I had included conferences with 1 team it would be confusing since there would be several conferences with only 1. With that being said I learned that the Big Ten had the most teams in the tournement with 9. I also discovered that the average seed from a multiple bid conference was about 7 according to the box plot. The data was interesting to interpret becuase the averages were somewhat misleading. This was becuase several conferences had large ranges compared to others. For example, the Big Ten had a range from 1 to 11 with 9 data points whereas the WCC had a range from 1 to 6 with 2 data points. Therefore, to decide who was the best seeded conference I decided to focus on the median which led me to believe that the best seeded conference was the Big Ten
